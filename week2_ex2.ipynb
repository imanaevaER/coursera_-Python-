```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import math

def f(p):
    return (math.sin(p/5)*np.exp(p/10)+5*np.exp(-p/2))


points = np.linspace(1, 15, 500)
y = list(map(f,points))
plt.plot(points, y)
```

**в случае многочлена первой степени:**

```python
a = np.array([[1, 1], [1, 15]])
b = np.array([f(1), f(15)])
x = np.linalg.solve(a, b)
print ("Решение системы:\n", x)
```

```python
def mn_1(p):
    return x[0]+x[1]*p

plt.plot(points, mn_1(points), points, list(map(f,points)))
```

**в случае многочлена второй степени:**

```python
a_2 = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
b_2 = np.array([f(1), f(8), f(15)])
x_2 = np.linalg.solve(a_2, b_2)
print ("Решение системы:\n", x_2)
```

```python
def mn_2(p):
    return x_2[0]+x_2[1]*p+x_2[2]*p**2

plt.plot(points, mn_2(points), points, list(map(f,points)))
```

**в случае многочлена третьей степени:**

```python
a_3 = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]])
b_3 = np.array([f(1), f(4), f(10), f(15)])
x_3 = np.linalg.solve(a_3, b_3)
print ("Решение системы:\n", x_3)
```

```python
def mn_3(p):
    return x_3[0]+x_3[1]*p+x_3[2]*p**2+x_3[3]*p**3

plt.plot(points, mn_3(points), points, list(map(f,points)))
```
