```python
import pandas as pd
import numpy as np
```

```python
file = open('sentences.txt')

data=[]
for line in file:
    l = line.strip()
    data.append(l)
    
file.close()
```

```python
# переводим в нижний регистр, делим на слова, удаляем пробелы
import re

all_symb = []
for i in data:
    all_symb.append(re.split('[^a-z]', i.lower()))


all_together = []
for i in all_symb:
    for j in i:
        all_together.append(j)


all_together.remove('')
```

```python
# создаем словарь уникальных слов
fin = set(all_together)
fin.remove('')
final = list(fin)

dict_words = { final[i] : i for i in range(0, (len(final)))}
```

```python
# заполняем матрицу

mat = np.zeros((len(data),len(final)))

for i in range(0, len(all_symb)):
    one_sent = all_symb[i]
    uniq_w = list(set(one_sent))
    for k in range(0, len(uniq_w)):
        if uniq_w[k] in dict_words:
            mat[i][dict_words[uniq_w[k]]] = one_sent.count(uniq_w[k])
            k+=1
        else:
            k+=1
```

```python
from scipy.spatial import distance

dist = []
for i in range(1, len(mat)):
    dist.append([i, distance.cosine(mat[0], mat[i])])
```

```python
pd.DataFrame(dist, columns=['sent', 'distance']).sort_values('distance')
```
